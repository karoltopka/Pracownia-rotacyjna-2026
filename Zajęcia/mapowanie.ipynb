```python
from rdkit import Chem
import pandas as pd

def can(smiles):
    try:
        mol = Chem.MolFromSmiles(str(smiles))
        return Chem.MolToSmiles(mol, canonical=True) if mol else None
    except:
        return None

fr_db = pd.read_excel("FR.xlsx")
raw_data = pd.read_csv("baza3.csv", sep="\t", encoding="utf-16")

fr_db["SMI_CAN"] = fr_db["SMILES"].apply(can)
raw_data["SMI_CAN"] = raw_data["SMILES"].apply(can)

match_cas = pd.merge(
    fr_db[["CAS", "ID"]],
    raw_data,
    left_on="CAS",
    right_on="CAS Number",
    how="right"
)

match_smiles = pd.merge(
    fr_db[["SMI_CAN", "ID"]],
    raw_data,
    on="SMI_CAN",
    how="right"
)

merged = match_cas.fillna(match_smiles)

final_output = merged.drop(columns=["CAS", "SMILES_CAN"], errors="ignore")
final_output.to_excel("WYNIK_zmapowany.xlsx", index=False)

print("Zakończono mapowanie")
```
