```python
import pandas as pd
from rdkit import Chem 

def canonicalize(smiles):
    if isinstance(smiles, str) and smiles.strip():
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            return Chem.MolToSmiles(mol, canonical=True)
    return None

df = pd.read_csv('./FR.txt', sep='\t')

df2 = pd.read_csv('./fire_rvgenetictoxicity.csv',sep='\t', encoding='utf-16',)



df.iloc[:, 3] = df.iloc[:, 3].apply(canonicalize)
df2.iloc[:, 3] = df2.iloc[:, 3].apply(canonicalize)

mapa_id2 = dict(zip(df.iloc[:, 3], df.iloc[:, 0]))


df2['FR_ID'] = df2.iloc[:, 3].map(mapa_id2)
print(df2.head)
df2.to_excel('wynik.xlsx', index=False)
```
